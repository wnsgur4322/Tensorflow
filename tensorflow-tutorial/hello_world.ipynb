```python
import keras
import numpy

model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer= 'sgd', loss='mean_squared_error')

xs= numpy.array([i-1 for i in range(7)], dtype=float)
ys= numpy.array([2*i-3 for i in range(7)], dtype=float)

model.fit(xs, ys, epochs=500)

print(model.predict([100.0]))

```

```python
import keras
import numpy as np

model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
xs = np.array([i for i in range(4)], dtype= float)
ys = np.array([50+50*i for i in range(4)], dtype= float)
model.fit(xs, ys, epochs = 1000)
print(model.predict([7.0]))
```
